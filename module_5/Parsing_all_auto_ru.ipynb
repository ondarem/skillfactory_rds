```python

dict_auto_ru = {} # Словарь, в который будут записываться объявления
j = 0 # кол-во объявлений
for year in range(2000, 2021):
    a = 1 #Переменная для перехода по страницам
    if year != 2020:
        section = "all"
    else:
        section = "used"
    while a <= 200: #Всего 99 страниц на сайте
        
        #Параметры запроса
        PARAMS = {
             'year_from': year, 'year_to': year, 
             'section': section,
             'category': "cars",
             'sort': "fresh_relevance_1-desc",
             'page': a
            }
        url = 'https://auto.ru/-/ajax/desktop/listing/' #URL на который будет отправлен запрос

        response = requests.post(url, json=PARAMS, headers=HEADERS) #Делаем post запрос на url
        data = response.json()['offers'] #Переменная data хранит полученные объявления

   
        # создаем словарь, в который будем записывать необходимые данные из каждого объявления
        dict_offers = {'body_type':0, 'brand':0, 'car_url':0, 'color':0, 'complectation_dict':0,
           'description':0, 'engine_displacement':0, 'engine_power':0, 'equipment_dict':0,
           'fuel_type':0, 'image':0, 'mileage':0, 'model_date':0, 'model_info':0, 'model_name':0,
           'name':0, 'number_ofdoors':0, 'parsing_unixtime':0, 'price_currency':0,
           'production_date':0, 'sell_id':0, 'super_gen':0, 'vehicle_configuration':0,
           'vehicle_transmission':0, 'vendor':0, 'owners':0, 'own':0, 'pts':0,
           'drive':0, 'rudder':0, 'condition':0, 'customs':0, 'mark':0, 'model':0, 'summary':0, 'price':0}

        i = 0 #Переменная для перехода по объявлениям
        while i <= len(data) - 1: #len(data)-1 это количество пришедших объявлений
            
            # Тип кузова
            try: dict_offers['body_type'] = data[i]['vehicle_info']['configuration']['human_name']
            except: dict_offers['body_type'] = None

            # Бренд
            try: dict_offers['brand'] = data[i]['vehicle_info']['mark_info']['name']
            except: dict_offers['brand'] = None
            
            # car_url
            dict_offers['car_url'] = None

            # Цвет автомобиля
            try: dict_offers['color'] = data[i]['color_hex']
            except: dict_offers['color'] = None

            # Комплектация
            try: dict_offers['complectation_dict'] = data[i]['vehicle_info']['complectation']
            except: dict_offers['complectation_dict'] = None

            # Описание автомобиля
            try: dict_offers['description'] = data[i]['description']
            except: dict_offers['description'] = None

            # Объем двигателя
            try: dict_offers['engine_displacement'] = data[i]['vehicle_info']['tech_param']['displacement']
            except: dict_offers['engine_displacement'] = None

            # Мощность двигателя
            try: dict_offers['engine_power'] = data[i]['vehicle_info']['tech_param']['power']
            except: dict_offers['engine_power'] = None

            # Доп комплектация
            try: dict_offers['equipment_dict'] = data[i]['vehicle_info']['equipment']
            except: dict_offers['equipment_dict'] = None

            # Тип используемого топлива
            try: dict_offers['fuel_type'] = data[i]['vehicle_info']['tech_param']['engine_type']
            except: dict_offers['fuel_type'] = None

            # Изображение
            dict_offers['image'] = None

            # Пробег автомобиля
            try: dict_offers['mileage'] = data[i]['state']['mileage']
            except: dict_offers['mileage'] = None

            # Год выпуска модели
            try: dict_offers['model_date'] = data[i]['vehicle_info']['super_gen']['year_from']
            except: dict_offers['model_date'] = None

            # Информация о модели
            try: dict_offers['model_info'] = data[i]['vehicle_info']['model_info']
            except: dict_offers['model_info'] = None

            # Название модели
            try: dict_offers['model_name'] =  data[i]['vehicle_info']['model_info']['code']
            except: dict_offers['model_name'] = None

            # Название
            try: dict_offers['name'] =  data[i]['vehicle_info']['tech_param']['human_name']
            except: dict_offers['name'] = None

            #Количество дверей у автомобиля
            try: dict_offers['number_ofdoors'] = data[i]['vehicle_info']['configuration']['doors_count']
            except: dict_offers['number_ofdoors'] = None

            # Время, когда был сделан парсинг
            try: dict_offers['parsing_unixtime'] = data[i]['additional_info']['fresh_date']
            except: dict_offers['parsing_unixtime'] = None

            # Валюта цены
            try: dict_offers['price_currency'] = data[i]['price_info']['currency']
            except: dict_offers['price_currency'] = None

            # Год выпуска автомобиля
            try: dict_offers['production_date'] = data[i]['documents']['year']
            except: dict_offers['production_date'] = None

            # ID продажи
            try: dict_offers['sell_id'] = data[i]['id']
            except: dict_offers['sell_id'] = None

            # super_gen
            try: dict_offers['super_gen'] = data[i]['vehicle_info']['super_gen']
            except: dict_offers['super_gen'] = None

            # vehicle_сonfiguration  
            try: dict_offers['vehicle_configuration'] = data[i]['vehicle_info']['configuration']
            except: dict_offers['vehicle_configuration'] = None

            # vehicle_transmission
            try: dict_offers['vehicle_transmission'] = data[i]['vehicle_info']['tech_param']['transmission']
            except: dict_offers['vehicle_transmission'] = None

            # Vendor
            try: dict_offers['vendor'] = data[i]['vehicle_info']['vendor']
            except: dict_offers['vendor'] = None

            # Количество владельцев автомобиля
            try: dict_offers['owners'] = data[i]['documents']['owners_number']
            except: dict_offers['owners'] = None

            # Владение
            try: dict_offers['own'] =  data[i]['documents']['purchase_date']
            except: dict_offers['own'] = None

            # ПТС автомобиля
            try: dict_offers['pts'] = data[i]['documents']['pts']
            except: dict_offers['pts'] = None

            # Привод
            try: dict_offers['drive'] = data[i]['vehicle_info']['tech_param']['gear_type']
            except: dict_offers['drive'] = None

            # Руль
            try: dict_offers['rudder'] = data[i]['vehicle_info']['steering_wheel']
            except: dict_offers['rudder'] = None

            # 'Состояние' - может быть state_not_beaten?
            try: dict_offers['condition'] = data[i]['state']['state_not_beaten']
            except: dict_offers['condition'] = None

            # Растаможен ли автомобиль (возвращает True или False)
            try: dict_offers['customs'] = data[i]['documents']['custom_cleared']
            except: dict_offers['customs'] = None

            # Марка автомобиля
            try: dict_offers['mark'] = data[i]['vehicle_info']['mark_info']['name']
            except: dict_offers['mark'] = None

            # Модель автомобиля
            try: dict_offers['model'] = data[i]['vehicle_info']['model_info']['name']
            except: dict_offers['model'] = None

            # Информация об автомобиле
            try: dict_offers['summary'] = data[i]['lk_summary']
            except: dict_offers['summary'] = None

            # Цена в рублях
            try: dict_offers['price'] = data[i]['price_info']['RUR']
            except: dict_offers['price'] = None

            
            dict_auto_ru[j] = dict_offers.copy()


            i += 1 # Увеличиваем переменную перехода по объявлениям на 1
            j += 1 # Увеличиваем переменную количества объявлений
            #sleep(randint(1,3))
        if i>0:
            print('Page: ' + str(a), year, 'кол-во объявлений', i, j) #Выводим сообщение, какая страница записалась
        a += 1 #Увеличиваем переменную страницы сайта на 1
    
    

# Записываем данные в файл
with open("dict_auto_ru.json", "w") as f:
    json.dump(dict_auto_ru, f)

print('Successfully', j) #Выводим информацию об успешном выполнении
# Page: 76 2018 кол-во объявлений 31 49259
# Successfully 49259 2008,2011,2013 - 99стр
# Page: 68 2019 кол-во объявлений 14 2497
# Successfully 2497
```
